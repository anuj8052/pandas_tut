```python
df = pd.read_csv("downloads/train.csv", parse_dates = ['day'])
df
```

```python
df.set_index('day', inplace = True)
df.head()
```

```python
# 1st method to fill the NaN values
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temp': 0,
    'windspeed': 0,
    'event': 'no_event'
})
```

```python
new_df
```

```python
new_df = df.fillna({
    'temp': df['temp'].mean()
})
new_df
```

```python
df.describe()
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.fillna(method= "bfill")
new_df
```

```python
# we can also use limit in while filling na values like 
# limit = 1 or 2 
# ex
# new_df = df.fillna(method = 'bfill', limit= 1)
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method = "time")
new_df
```
