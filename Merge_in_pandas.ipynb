```python
# basic merge using a dataframe column
```

```python
df = pd.DataFrame({
    "city":["delhi", "mumbai", "banglore"],
    "temp":[43,34,47]
})
df
```

```python
df1 = pd.DataFrame({
    "city":["delhi", "mumbai", "banglore"],
    "temp":[43, 32, 45]
})
df1
```

```python
df3 = pd.merge(df, df1, on = "city")
df3
```

```python
# there are four type of joins in database
# left, right, full, inner join
```

```python
df1 = pd.DataFrame({
    "city":["mumbai", "kharagpur", "kolkata"], 
    "hum":[43,45,56]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["mumbai", "kharagpur", "agra","patna"],
    "hum":[43, 34, 45, 45]
})
df2
```

```python
df4 = pd.merge(df1, df2, on = "city", how = "inner")
df4
# inner join = intersection between two cities
```

```python
df5 = pd.merge(df1, df2, on = "city", how = "outer")
df5
# outer = full join = union
```

```python
df6 = pd.merge(df1, df2, on = "city", how = "left")
df6
```

```python
df7 = pd.merge(df1, df2, on = "city", how = "right")
df7
```

```python
# indicator flag
```

```python
d1 = pd.merge(df1, df2, on = "city", how = "outer", indicator=True)
d1
```

```python
# suffixes
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1, df2, on = "city", how = "outer", 
               suffixes = ('_first', '_sec'))
df3
```

```python
# join
```

```python
df1 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperatur": [21,14,35],
    "humidity": [65,68,71]
})
df1.set_index("city", inplace = True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "temperatur": [65,68,75],
    "humidity": [35,48,21]
})
df2.set_index("city", inplace = True)
df2
```

```python
df1.join(df2, lsuffix = '_l', rsuffix = '_r')
```
