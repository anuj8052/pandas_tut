```python
data = pd.read_csv("downloads/Anuj_train.csv")
data
```

```python
data = data.drop(['Unnamed: 4', 'Unnamed: 5','Unnamed: 6'], axis = 1)
```

```python
# columns() to Convert Row to Column Header. You can use df. columns=df. iloc[0] 
# to set the column labels by extracting the first row
```

```python
data.columns = data.iloc[0]
data
```

```python
# this given querry tell how we can drop a row 
```

```python
new_data = data.drop(data.index[0])
```

```python
new_data
```

```python
# we use regx to find the pattern and then we replace that type of pattern
```

```python
new_data1 = new_data.replace({
    'temp': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},'', regex = True)
```

```python
new_data1
```

```python
new_data2 = new_data1.replace({
    'temp':'-9999',
    'windspeed':'-9999',
    'event':'0'
}, np.NaN)
```
